#  Model Hamiltonian #

We start with the following Hamiltonian:
\begin{align}
  \mathcal{H} = &- \sum_{< i,j >,\sigma,m} t^{m} (c^{\dagger}_{i,m,\sigma}c_{j,m,\sigma} + h.c) \nonumber \\ 
 & + U \sum_{i,m} n_{i,m,\uparrow}n_{i,m,\downarrow} + (U-2J)\sum_{m\neq l,i} n_{i,m,\uparrow}n_{i,l,\downarrow} \nonumber \\
 & +(U-3J) \sum_{m < l,i,\sigma}n_{i,m,\sigma}n_{i,l,\sigma}\quad.
\end{align}


Our goal is to study the model in a mean field fashion, allowing for tree different order parameters in the Hartree-Fock Hamiltonian.
In particular our variational parameters are:
  - $\Delta_{CDW} = \frac{1}{N}\sum_{i,m} (-1)^{R_i} n_{i,m}$ ,

  - $\Delta_{m} = \langle c^{\dagger}_{i,m,\uparrow} c^{\dagger}_{i,m,\downarrow}\rangle $ ,
  
  - $\Delta_{m,n} = \langle c^{\dagger}_{i,m,\uparrow} c^{\dagger}_{i,n,\downarrow}\rangle $.


  We first write the free particle dispersion in momentum space, we further assume to be in two dimensions and in a square lattice.
  Hence, the hopping term can be written as:

  
  \begin{align}
    \mathcal{H}_{t} = -  2\sum_{\bm{k},m,\sigma} t^{m} (cos(k_x a) + cos(k_y a)) c^{\dagger}_{\bm{k},m,\sigma}c_{\bm{k},m,\sigma}.
  \end{align}
Where 'a' is the lattice constant.
We start rewriting the Hamiltonian in the Hartree Fock approximation by noticing that we can write:

\begin{align}
 n_{i,m,\sigma} = \frac{1}{2}(1+ (-1)^{R_i} \Delta_{CDW}/2)
\end{align}


Hence the first term of the Hamiltonian is:

\begin{align}
\mathcal{H}_{U} = & \sum_{i,m,\sigma} \frac{1}{2}(1+\frac{(-1)^{R_i}}{2}\Delta_{CDW}) c^{\dagger}_{i,m,\sigma}c_{i,m,\sigma} - \frac{1}{2}\sum_{i}(1+\Delta_{CDW}^2/4)\nonumber\\
& -\sum_{i,m} \Delta_{m} c_{i,m,\uparrow}c_{i,m,\downarrow} +\sum_{i,m} \Delta^{*}_{m} c^{\dagger}_{i,m,\uparrow}c^{\dagger}_{i,\downarrow} \nonumber \\
&- \sum_{i,m} \Delta_{m} \Delta^{*}_{m}

\end{align}


And the second is:

\begin{align}
\mathcal{H}_{U-2J} = & \sum_{i,m,\sigma} \frac{1}{2}(1+\frac{(-1)^{R_i}}{2}\Delta_{CDW}) c^{\dagger}_{i,m,\sigma}c_{i,m,\sigma} - \frac{1}{2}\sum_{i}(1+\Delta_{CDW}^2/4)\nonumber\\
&-\sum_{i,m\neq n} \Delta_{m,n} c_{i,m,\uparrow}c_{i,n,\downarrow} +\sum_{i,m\neq n} \Delta^{*}_{m,n} c^{\dagger}_{i,m,\uparrow}c^{\dagger}_{i,n,\downarrow} \nonumber \\
&- \sum_{i,m\neq n} \Delta_{m,n} \Delta^{*}_{m,n}
\end{align}

Finally the last term is:

\begin{align}
  \mathcal{H}_{U-3J} = \sum_{m,i, \sigma} \frac{1}{2} (1 + (-1)^{\bm{R}_i} \frac{\Delta}{2}) c^{\dagger}_{i,n,\sigma} c_{i,n,\sigma} - \frac{1}{2}\sum_{i} (1 + \frac{\Delta^2}{4}) 
\end{align}

We can put all the term together to obtain the following Hamiltonian:

\begin{align}
    \mathcal{H} = &-  2\sum_{\bm{k},m,\sigma} t^{m} (cos(k_x a) + cos(k_y a)) c^{\dagger}_{\bm{k},m,\sigma}c_{\bm{k},m,\sigma} \nonumber \\
    & +(3U-5J)\sum_{i,m,\sigma} \frac{1}{2}(1+\frac{(-1)^{R_i}}{2}\Delta_{CDW}) c^{\dagger}_{i,m,\sigma}c_{i,m,\sigma} \nonumber \\
    & -U (\sum_{i,m} \Delta_{m} c_{i,m,\uparrow}c_{i,m,\downarrow} -\sum_{i,m} \Delta^{*}_{m} c^{\dagger}_{i,m,\uparrow}c^{\dagger}_{i,\downarrow} )\nonumber \\
    & -(U-2J)(\sum_{i,m\neq n} \Delta_{m,n} c_{i,m,\uparrow}c_{i,n,\downarrow} -\sum_{i,m\neq n} \Delta^{*}_{m,n} c^{\dagger}_{i,m,\uparrow}c^{\dagger}_{i,n,\downarrow}) \nonumber \\
    & - \sum_i \Bigg[ \frac{3U-5J}{2} (1+ \frac{\Delta^2}{4}) +U \sum_{m} |\Delta_m|^2 + (U-2J)\sum_{m\neq n}|\Delta_{m,n}|^2\Bigg]

\end{align}

In Fourier space the Hamiltonian reads:

\begin{align}
    \mathcal{H} = &-  2\sum_{\bm{k},m,\sigma} t^{m} (cos(k_x a) + cos(k_y a)) c^{\dagger}_{\bm{k},m,\sigma}c_{\bm{k},m,\sigma} \nonumber \\
    & +\frac{(3U-5J)}{2}\sum_{k,m,\sigma} ( c^{\dagger}_{k,m,\sigma}c_{k,m,\sigma}+\frac{1}{2}\Delta_{CDW}c^{\dagger}_{k+ Q,m,\sigma}c_{k,m,\sigma})\nonumber \\
    & -U (\sum_{k,m} \Delta_{m} c_{k,m,\uparrow}c_{-k,m,\downarrow} -\sum_{k,m} \Delta^{*}_{m} c^{\dagger}_{k,m,\uparrow}c^{\dagger}_{-k,\downarrow} )\nonumber \\
    & -(U-2J)(\sum_{k,m\neq n} \Delta_{m,n} c_{k,m,\uparrow}c_{-k,n,\downarrow} -\sum_{k,m\neq n} \Delta^{*}_{m,n} c^{\dagger}_{k,m,\uparrow}c^{\dagger}_{-k,n,\downarrow}) \nonumber \\
    & - \sum_k \Bigg[ \frac{3U-5J}{2} (1+ \frac{\Delta_{CDW}^2}{4}) +U \sum_{m} |\Delta_m|^2 + (U-2J)\sum_{m\neq n}|\Delta_{m,n}|^2\Bigg]

\end{align}
